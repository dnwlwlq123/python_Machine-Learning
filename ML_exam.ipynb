```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic_group = titanic.groupby("survived")
titanic_group[["age", "fare"]].mean()
titanic_group[["age", "fare"]].std()
```

```python
sns.regplot(titanic,
            x = "age",
            y = "fare",
            fit_reg=True,
            color = "red",
            order = 1,
            label = "Titanic")
plt.title("나이 vs 요금")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.lmplot(penguins,
           x = "bill_length_mm",
           y = "bill_depth_mm",
           hue = "species",
           col = "sex")
plt.show()
```
