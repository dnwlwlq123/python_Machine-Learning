```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score 

```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
titanic.info()
```

```python
titanic = titanic.loc[:, ["survived", "pclass", "sex", "age", "sibsp", "parch", "fare"]]
titanic.dropna(subset=["age"], axis=0, inplace=True)

# 범주형 데이터 변환
```

```python
X = titanic.loc[:, "pclass":"fare"]
y = titanic["survived"]

```

```python
##DT, rf, gb survived = 1 alive
X = pd.get_dummies(X, columns = ["sex"], drop_first=True)
print(X.head())
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate, StratifiedKFold, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
## dt
dt = DecisionTreeClassifier()
params = {
    "max_depth" : range(3, 20, 1),
    "max_leaf_nodes" : range(3, 20, 1) 
}
splitter = StratifiedKFold(n_splits = 5, shuffle = True)
gs_dt = GridSearchCV(dt, param_grid = params, cv = splitter)
gs_dt.fit(X,y)
```

```python
print("best parameters = ",gs_dt.best_estimator_)
print("best cv score = {:.4f}".format( gs_dt.best_score_))
```

```python
## rf
rf = RandomForestClassifier()

params = {
    "max_depth" : range(3, 15, 1),
    "n_estimators" : range(90, 120, 1)
}
splitter = StratifiedKFold(n_splits=5, shuffle=True)
grid_rf = GridSearchCV(rf, param_grid=params, cv = splitter)
grid_rf.fit(X, y)
```

```python
print("best parameters = ",grid_rf.best_estimator_)
print("best cv score = {:.4f}".format( grid_rf.best_score_))
```

```python
gb = GradientBoostingClassifier()
params = {
    "max_depth" : range(3, 20, 1),
    "learning_rate": np.arange(0.1, 1.0, 0.1)
}
splitter = StratifiedKFold(n_splits=5, shuffle=True)
gs_gb = GridSearchCV(gb, param_grid=params, cv = splitter)
gs_gb.fit(X,y)
```

```python
print("best parameters = ",gs_gb.best_estimator_)
print("best cv score = {:.4f}".format( gs_gb.best_score_))
```

```python
wine = pd.read_csv("https://bit.ly/wine-date")
wine.head()
```

```python
wine["class"] = wine["class"].astype("int32").astype('category')
wine.head()
```

```python
X = wine.iloc[:, :-1]
y = wine["class"]
```

```python
standardscaler = StandardScaler()
X_scaled = standardscaler.fit_transform(X)
X_scaled[:10]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=100)
splitter = StratifiedKFold(n_splits=5, shuffle=True)
scores = cross_validate(lr, X_scaled, y, cv = splitter)
```

```python
print(scores["test_score"])
```

```python
lr.fit(X_scaled, y)
```
